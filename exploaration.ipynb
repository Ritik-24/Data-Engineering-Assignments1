```python
import numpy as np

df['quantity'] = np.random.randint(1, 6, size=len(df))
df['total_amount'] = df['initial_price'] * df['quantity']
df[['title', 'initial_price', 'quantity', 'total_amount']]
```

```python
df['final_price'] = df['final_price'].astype(str).str.replace(r'[^\d.]', '', regex=True)
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')
df.to_csv('cleaned_dataset.csv', index=False, encoding='utf-8-sig')
df[['title', 'initial_price', 'final_price']]
```

```python
category_summary = df.groupby('category')['total_amount'].sum()
category_summary
```

```python
import re

def clear_corrupted_blocks(text):
    if isinstance(text, str):
        cleaned = re.sub(r'[^\x00-\x7F]+', '', text)
        cleaned = re.sub(r'[\x00-\x1F\x7F]', '', cleaned)
        return cleaned.strip()
    return text

columns_to_clean = ['seller_name', 'seller_info', 'seller_information']

for col in columns_to_clean:
    if col in df.columns:
        df[col] = df[col].apply(clear_corrupted_blocks)

df.to_csv('cleaned_dataset.csv', index=False, encoding='utf-8-sig')
```
